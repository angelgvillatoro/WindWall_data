```python
# Create organized structure for plotting
plot_data = {}
for pwm in rpm_data:
    plot_data[pwm] = {
        "time": [point["time"] for point in rpm_data[pwm]],
        "rpm": [point["rpm"] for point in rpm_data[pwm]],
        "pulses": [point["pulses"] for point in rpm_data[pwm]]
    }

# PWM values sorted for consistent plotting
pwm_values = sorted(rpm_data.keys(), reverse=True)
```

```python
# Plotting
import matplotlib.pyplot as plt
import numpy as np

# Set up the plot
plt.figure(figsize=(14, 10))

# Create a color map for different PWM values
colors = plt.cm.viridis(np.linspace(0, 1, len(pwm_values)))

# Plot RPM vs Time for each PWM value
for i, pwm in enumerate(pwm_values):
    plt.plot(plot_data[pwm]["time"], plot_data[pwm]["rpm"], 
             label=f'PWM = {pwm}', color=colors[i], linewidth=2, marker='o', markersize=4)

plt.xlabel('Time (seconds)', fontsize=12)
plt.ylabel('RPM', fontsize=12)
plt.title('Motor RPM vs Time for Different PWM Values', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.legend(title='PWM Value', fontsize=10, title_fontsize=11)
plt.tight_layout()
plt.show()
```

```python
# Create subplots for better comparison
fig, axes = plt.subplots(2, 4, figsize=(16, 10))
axes = axes.flatten()

for i, pwm in enumerate(pwm_values):
    ax = axes[i]
    ax.plot(plot_data[pwm]["time"], plot_data[pwm]["rpm"], 
            color=colors[i], linewidth=2, marker='o', markersize=4)
    ax.set_title(f'PWM = {pwm}', fontsize=11, fontweight='bold')
    ax.set_xlabel('Time (s)')
    ax.set_ylabel('RPM')
    ax.grid(True, alpha=0.3)
    ax.set_ylim(0, 8000)  # Consistent y-axis for comparison

# Hide empty subplot if we have less than 8 PWM values
for i in range(len(pwm_values), len(axes)):
    axes[i].set_visible(False)

plt.suptitle('Motor RPM Response at Different PWM Values', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Create a summary plot of maximum RPM vs PWM
max_rpms = []
for pwm in pwm_values:
    max_rpms.append(max(plot_data[pwm]["rpm"]))

plt.figure(figsize=(10, 6))
plt.plot(pwm_values, max_rpms, 'o-', linewidth=2, markersize=8, color='darkred')
plt.xlabel('PWM Value', fontsize=12)
plt.ylabel('Maximum RPM', fontsize=12)
plt.title('Maximum RPM Achieved vs PWM Value', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)

# Add annotations for each point
for i, (pwm, rpm) in enumerate(zip(pwm_values, max_rpms)):
    plt.annotate(f'{rpm:.0f}', (pwm, rpm), textcoords="offset points", 
                 xytext=(0,10), ha='center', fontsize=9)

plt.tight_layout()
plt.show()
```
