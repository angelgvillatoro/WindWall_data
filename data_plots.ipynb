```python
import numpy as np
import pandas as pd

PWMS = [1200, 1300, 1400, 1500, 1600, 1800, 2000]

# rpm_data -> df
rows = []
for pwm_cmd in PWMS:
    for r in rpm_data[pwm_cmd]:
        rows.append({"pwm_cmd": pwm_cmd,
                     "time": float(r["time"]),
                     "rpm": float(r["rpm"])})
df = pd.DataFrame(rows).sort_values(["pwm_cmd", "time"]).reset_index(drop=True)

def estimate_steady_rpm_per_pwm(df, pwms, min_time_after_start=0.25, drop_frac=0.85, tail_points=5):
    steady = {}
    for pwm_cmd in pwms:
        g = df[df["pwm_cmd"] == pwm_cmd].sort_values("time").reset_index(drop=True)
        g2 = g[g["time"] >= min_time_after_start].copy()
        if len(g2) < tail_points:
            g2 = g.copy()

        rpm_vals = g2["rpm"].values
        t_vals = g2["time"].values

        i_peak = int(np.argmax(rpm_vals))
        rpm_peak = float(rpm_vals[i_peak])

        t_drop = None
        for i in range(i_peak + 1, len(g2)):
            if rpm_vals[i] < drop_frac * rpm_peak:
                t_drop = float(t_vals[i])
                break

        if t_drop is None:
            plateau = g2.tail(tail_points)
        else:
            w = g2[g2["time"] < t_drop]
            plateau = w.tail(tail_points) if len(w) >= tail_points else g2.tail(tail_points)

        steady[pwm_cmd] = float(np.median(plateau["rpm"].values))

    return pd.Series(steady, name="rpm_steady").sort_index()

steady_by_pwm = estimate_steady_rpm_per_pwm(df, PWMS)
print(steady_by_pwm)
```

```python
import numpy as np
import pandas as pd

def crossing_time_first(t, y, target, direction="up"):
    t = np.asarray(t); y = np.asarray(y)
    cond = (y >= target) if direction == "up" else (y <= target)
    idxs = np.where(cond)[0]
    if len(idxs) == 0:
        return np.nan
    j = int(idxs[0])
    if j == 0:
        return float(t[0])
    t1, t2 = float(t[j-1]), float(t[j])
    y1, y2 = float(y[j-1]), float(y[j])
    if y2 == y1:
        return float(t2)
    alpha = (target - y1) / (y2 - y1)
    return float(t1 + alpha * (t2 - t1))

def estimate_between_pwms_using_high_test(df, target_map, pwm_high, pwm_low):
    """
    Estima:
      - accel low->high: cruce(target_low) -> cruce(target_high) en tramo de subida del ensayo high
      - decel high->low: cruce(target_high) -> cruce(target_low) en tramo de bajada del ensayo high
    target_map: Serie indexada por PWM con rpm objetivo (aqu: rpm_peak)
    """
    g = df[df["pwm_cmd"] == pwm_high].sort_values("time").reset_index(drop=True)
    t = g["time"].values
    rpm = g["rpm"].values

    target_high = float(target_map.loc[pwm_high])
    target_low  = float(target_map.loc[pwm_low])

    i_peak = int(np.argmax(rpm))
    t_rise, rpm_rise = t[:i_peak+1], rpm[:i_peak+1]
    t_fall, rpm_fall = t[i_peak:], rpm[i_peak:]

    t_low_rise  = crossing_time_first(t_rise, rpm_rise, target_low,  direction="up")
    t_high_rise = crossing_time_first(t_rise, rpm_rise, target_high, direction="up")

    t_high_fall = crossing_time_first(t_fall, rpm_fall, target_high, direction="down")
    t_low_fall  = crossing_time_first(t_fall, rpm_fall, target_low,  direction="down")

    t_acc = (t_high_rise - t_low_rise) if np.isfinite(t_low_rise) and np.isfinite(t_high_rise) else np.nan
    t_dec = (t_low_fall - t_high_fall) if np.isfinite(t_high_fall) and np.isfinite(t_low_fall) else np.nan

    return {
        "pwm_low": pwm_low,
        "pwm_high": pwm_high,
        "rpm_low_target": target_low,
        "rpm_high_target": target_high,
        "t_acc_low_to_high_s": t_acc,
        "t_dec_high_to_low_s": t_dec,
    }

# targets = rpm_peak por PWM
target_map = peak_by_pwm["rpm_peak"]

# Ejemplo: 1800 <-> 2000 usando el ensayo de 2000
print(estimate_between_pwms_using_high_test(df, target_map, pwm_high=2000, pwm_low=1800))

# Todas las parejas
PWMS = [1200, 1300, 1400, 1500, 1600, 1800, 2000]
out = []
for high in PWMS:
    for low in PWMS:
        if low >= high: 
            continue
        out.append(estimate_between_pwms_using_high_test(df, target_map, pwm_high=high, pwm_low=low))
pairs_peak = pd.DataFrame(out).sort_values(["pwm_high","pwm_low"])
pairs_peak
```

```python
import numpy as np
import pandas as pd

# --- Inputs que ya tienes ---
# pairs_peak: DataFrame con columnas
# ['pwm_low','pwm_high','t_acc_low_to_high_s','t_dec_high_to_low_s', ...]
# PWMS: lista [1200, 1300, 1400, 1500, 1600, 1800, 2000]

PWMS = [1200, 1300, 1400, 1500, 1600, 1800, 2000]
PWM_MAX = max(PWMS)

# Etiquetas en porcentaje (respecto a PWM_MAX)
labels = [f"{int(round(100*p/PWM_MAX))}%" for p in PWMS]

# Matriz NxN con NaN, diagonal a 0
N = len(PWMS)
M = pd.DataFrame(np.nan, index=labels, columns=labels)
np.fill_diagonal(M.values, 0.0)

# Map PWM -> label %
pwm_to_label = {p: f"{int(round(100*p/PWM_MAX))}%" for p in PWMS}

# Rellenar:
# - por encima de diagonal (low->high): aceleracin
# - por debajo (high->low): desaceleracin
for _, r in pairs_peak.iterrows():
    low = int(r["pwm_low"])
    high = int(r["pwm_high"])

    i = pwm_to_label[low]   # fila low
    j = pwm_to_label[high]  # columna high

    # Tringulo superior: aceleracin low->high
    M.loc[i, j] = float(r["t_acc_low_to_high_s"])

    # Tringulo inferior: desaceleracin high->low
    M.loc[j, i] = float(r["t_dec_high_to_low_s"])

# Asegurar orden vertical 1200->2000 y horizontal 1200->2000
M = M.reindex(index=labels, columns=labels)
M.round(2)

```

```python
import numpy as np
import matplotlib.pyplot as plt

# M es tu DataFrame final (ndices y columnas en %)
A = M.to_numpy(dtype=float)
labels = list(M.index)

fig, ax = plt.subplots(figsize=(7, 6))

im = ax.imshow(A, aspect="equal")

# Ejes
ax.set_xticks(range(len(labels)))
ax.set_yticks(range(len(labels)))
ax.set_xticklabels(labels, rotation=45, ha="right")
ax.set_yticklabels(labels)

ax.set_xlabel("PWM [% of max]")
ax.set_ylabel("PWM [% of max]")
ax.set_title("Acceleration (upper) and Deceleration (lower) times")

# Anotar valores (2 decimales)
for i in range(len(labels)):
    for j in range(len(labels)):
        if np.isfinite(A[i, j]):
            ax.text(j, i, f"{A[i, j]:.2f}", ha="center", va="center")

# Colorbar
cbar = fig.colorbar(im, ax=ax)
cbar.set_label("Time [s]")

plt.tight_layout()
plt.show()
```
