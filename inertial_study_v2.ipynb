```python
import re
import numpy as np
import pandas as pd

# =========================================================
# 1) PARSE: raw log -> df
# =========================================================
def parse_rpm_log(raw_text: str) -> pd.DataFrame:
    lines = raw_text.splitlines()

    rows = []
    for ln in lines:
        ln = ln.strip()
        if not ln:
            continue

        # timestamp del log: I (822330) ...
        m_stamp = re.search(r"I\s*\((\d+)\)", ln)
        stamp_ms = int(m_stamp.group(1)) if m_stamp else np.nan

        # eventos start/stop
        event = None
        if "Motor started" in ln:
            event = "start"
        elif "Motor stopped" in ln and "Pulses:" not in ln:
            event = "stop"

        # mediciones tipo: Pulses: 5, RPM: 1500.00, Time: 0.400 seconds
        m_meas = re.search(r"Pulses:\s*([0-9]+)\s*,\s*RPM:\s*([0-9.]+)", ln)
        m_time = re.search(r"Time:\s*([0-9.]+)\s*seconds", ln)

        if m_meas:
            pulses = int(m_meas.group(1))
            rpm = float(m_meas.group(2))
            time_s = float(m_time.group(1)) if m_time else np.nan
            rows.append(
                dict(
                    stamp_ms=stamp_ms,
                    event="meas",
                    pulses=pulses,
                    rpm=rpm,
                    time_s=time_s,
                )
            )
        elif event is not None:
            rows.append(dict(stamp_ms=stamp_ms, event=event, pulses=np.nan, rpm=np.nan, time_s=np.nan))

    df = pd.DataFrame(rows).sort_values("stamp_ms").reset_index(drop=True)

    # running flag: entre 'start' y 'stop'
    df["running"] = False
    running = False
    for i, ev in enumerate(df["event"].values):
        if ev == "start":
            running = True
        elif ev == "stop":
            running = False
        df.at[i, "running"] = running if ev != "stop" else False  # stop marca motor parado

    return df


# =========================================================
# 2) BUILD RUNS: df -> meas(with run_id) + run_summary
# =========================================================
def build_runs(df: pd.DataFrame, min_run_s: float = 0.3, idle_gap_ms: int = 3000) -> tuple[pd.DataFrame, pd.DataFrame]:
    """
    Detecta runs usando solo mediciones:
    - Una run empieza cuando pasamos de rpm==0 a rpm>0 (o tras un gap largo)
    - Termina cuando rpm==0 y se mantiene, o hay gap largo sin datos
    """
    meas = df[df["event"].eq("meas")].copy()
    meas = meas.sort_values("stamp_ms").reset_index(drop=True)

    # motor "girando": rpm > 0 (umbral)
    meas["is_on"] = meas["rpm"] > 0.0

    # gap entre muestras
    meas["dt_ms"] = meas["stamp_ms"].diff()

    # nuevo run si:
    #  - first on
    #  - transicin off->on
    #  - gap largo
    new_run = (
        meas["is_on"] & (
            meas["is_on"].shift(1, fill_value=False).eq(False) |
            (meas["dt_ms"].fillna(0) > idle_gap_ms)
        )
    )

    meas["run_id"] = new_run.cumsum()
    # quitamos puntos "off" (rpm==0) de las runs, porque nos interesan solo tramos activos
    meas_on = meas[meas["is_on"]].copy()

    # reindex run_id para que empiece en 1..N sin huecos
    rid_map = {rid:i+1 for i, rid in enumerate(sorted(meas_on["run_id"].unique()))}
    meas_on["run_id"] = meas_on["run_id"].map(rid_map).astype(int)

    # resumen por run
    g = meas_on.groupby("run_id", sort=True)
    run_summary = g.agg(
        n=("rpm", "size"),
        stamp_start=("stamp_ms", "min"),
        stamp_end=("stamp_ms", "max"),
        dur_s_abs=("stamp_ms", lambda x: (x.max() - x.min())/1000.0),
        rpm_mean=("rpm", "mean"),
        rpm_std=("rpm", "std"),
        rpm_min=("rpm", "min"),
        rpm_max=("rpm", "max"),
    ).reset_index()

    run_summary["rpm_range"] = run_summary["rpm_max"] - run_summary["rpm_min"]

    # filtro mnimo duracin
    run_summary = run_summary[run_summary["dur_s_abs"] >= float(min_run_s)].copy()
    keep = set(run_summary["run_id"].tolist())
    meas_on = meas_on[meas_on["run_id"].isin(keep)].copy()

    return meas_on.reset_index(drop=True), run_summary.reset_index(drop=True)


# =========================================================
# 3) PWM mapping (tu tabla) + dinmica por run
# =========================================================
PLATEAU_MAP = {
    1200: 1200.000,
    1300: 2400.000,
    1400: 3000.000,
    1500: 4199.915,
    1600: 5100.000,
    1800: 6600.000,
    2000: 7500.000,
}
PWM_ORDER = [1200, 1300, 1400, 1500, 1600, 1800, 2000]
plateau_df = pd.DataFrame({"pwm": list(PLATEAU_MAP.keys()),
                           "rpm_plateau": list(PLATEAU_MAP.values())}).sort_values("pwm").reset_index(drop=True)

def _smooth_rpm(rpm, win=5):
    return pd.Series(rpm).rolling(win, center=True, min_periods=1).median().to_numpy()

def _crossing_time(t, y, thr):
    t = np.asarray(t, float); y = np.asarray(y, float)
    if len(t) < 2:
        return np.nan
    mask = (y[:-1] < thr) & (y[1:] >= thr)
    idx = np.where(mask)[0]
    if len(idx) == 0:
        return np.nan
    i = idx[0]
    t0, t1 = t[i], t[i+1]
    y0, y1 = y[i], y[i+1]
    if y1 == y0:
        return t1
    return t0 + (thr - y0) * (t1 - t0) / (y1 - y0)

def assign_pwm_by_nearest_plateau(rpm_ref):
    vals = plateau_df["rpm_plateau"].to_numpy()
    j = int(np.argmin(np.abs(vals - rpm_ref)))
    return int(plateau_df.loc[j, "pwm"]), float(plateau_df.loc[j, "rpm_plateau"])

def analyze_one_run(run_df):
    # usa tiempo por stamp (robusto, porque tu Time se resetea)
    t = (run_df["stamp_ms"].to_numpy(float) - run_df["stamp_ms"].iloc[0]) / 1000.0
    rpm = _smooth_rpm(run_df["rpm"].to_numpy(float), win=5)

    rpm_ref = float(np.nanquantile(rpm, 0.95))
    pwm, target = assign_pwm_by_nearest_plateau(rpm_ref)

    t10 = _crossing_time(t, rpm, 0.10 * target)
    t90 = _crossing_time(t, rpm, 0.90 * target)
    tr = (t90 - t10) if (np.isfinite(t10) and np.isfinite(t90)) else np.nan

    rpm_peak = float(np.nanmax(rpm))
    overshoot = (rpm_peak - target) / target if target > 0 else np.nan

    return dict(
        run_id=int(run_df["run_id"].iloc[0]),
        pwm_assigned=pwm,
        target_rpm=target,
        rpm_ref_p95=rpm_ref,
        dur_s=float(t.max()),
        t10_s=float(t10) if np.isfinite(t10) else np.nan,
        t90_s=float(t90) if np.isfinite(t90) else np.nan,
        tr_10_90_s=float(tr) if np.isfinite(tr) else np.nan,
        rpm_peak=rpm_peak,
        overshoot_rel=float(overshoot) if np.isfinite(overshoot) else np.nan,
    )


# =========================================================
# PIPELINE: raw -> df -> meas/run_summary -> dyn/summary_pwm
# =========================================================
if "raw" not in globals():
    raise NameError("No encuentro la variable 'raw'. Define raw como string (triple comillas) en una celda anterior.")

df = parse_rpm_log(raw)

meas, run_summary = build_runs(df, min_run_s=0.3, idle_gap_ms=3000)

print("runs detectadas:", run_summary["run_id"].nunique())
display(run_summary.sort_values("run_id").head(10))

# dinmica por run
dyn = pd.DataFrame([analyze_one_run(g) for _, g in meas.groupby("run_id", sort=True)])

dyn["pwm_rank"] = dyn["pwm_assigned"].map({p:i for i,p in enumerate(PWM_ORDER)}).fillna(999).astype(int)
dyn = dyn.sort_values(["pwm_rank", "run_id"]).drop(columns=["pwm_rank"]).reset_index(drop=True)

display(dyn)

print("\nConteo de runs por PWM asignado:")
display(dyn["pwm_assigned"].value_counts().sort_index())

summary_pwm = dyn.groupby("pwm_assigned").agg(
    n_runs=("run_id", "count"),
    target_rpm=("target_rpm", "first"),
    tr_10_90_med=("tr_10_90_s", "median"),
    t90_med=("t90_s", "median"),
    overshoot_med=("overshoot_rel", "median"),
    dur_med=("dur_s", "median"),
).reset_index().sort_values("pwm_assigned")

print("\nResumen por PWM:")
display(summary_pwm)
```

```python
import numpy as np

def first_crossing_time(t, rpm, level, direction="up", hold=3):
    """
    t, rpm: arrays 1D
    level: rpm a cruzar
    direction: 'up' o 'down'
    hold: n mnimo de muestras consecutivas tras el cruce
    """
    t = np.asarray(t)
    rpm = np.asarray(rpm)

    if direction == "up":
        crossed = (rpm[:-1] < level) & (rpm[1:] >= level)
        ok_side = rpm >= level
    else:
        crossed = (rpm[:-1] > level) & (rpm[1:] <= level)
        ok_side = rpm <= level

    idx = np.where(crossed)[0]

    for i in idx:
        j0 = i + 1
        j1 = min(len(rpm), j0 + hold)

        # confirmacin: se queda del lado correcto
        if np.all(ok_side[j0:j1]):
            t0, t1 = t[i], t[i+1]
            y0, y1 = rpm[i], rpm[i+1]

            if y1 == y0:
                return t1  # fallback seguro

            # interpolacin lineal
            return t0 + (level - y0) * (t1 - t0) / (y1 - y0)

    return np.nan
```

```python
import pandas as pd

def crossing_times_per_run(run_df, levels):
    """
    Devuelve:
      t_up[level] = tiempo de cruce en aceleracin
      t_dn[level] = tiempo de cruce en deceleracin
    """
    r = run_df.sort_values("stamp_ms").copy()

    # tiempo real relativo al inicio del run
    t = (r["stamp_ms"].to_numpy() - r["stamp_ms"].iloc[0]) / 1000.0
    rpm = (
        r["rpm"]
        .rolling(5, center=True, min_periods=1)
        .median()
        .to_numpy()
    )

    t_up = {}
    t_dn = {}

    for lvl in levels:
        t_up[lvl] = first_crossing_time(t, rpm, lvl, direction="up", hold=3)
        t_dn[lvl] = first_crossing_time(t, rpm, lvl, direction="down", hold=3)

    return t_up, t_dn
```

```python
import numpy as np
import pandas as pd

# ============================
# CONFIG
# ============================
PWM_COL = "pwm"
RUN_COL = "run_id"
T_COL   = "t_rel_s"
RPM_COL = "rpm_f" if "rpm_f" in m.columns else "rpm"

# ventana del plateau: usamos el 20%-80% del tiempo total (evita accel/decel)
PLATEAU_A = 0.2
PLATEAU_B = 0.8

# redondeo para hacer robusto el modo (ajusta: 25, 50, 100...)
ROUND_TO = 50

# si hay pocos puntos en ventana, hacemos fallback
MIN_PTS_WINDOW = 5

# ============================
# helper: modo robusto en el plateau
# ============================
def plateau_mode_rpm(g, round_to=50, a=0.2, b=0.8, min_pts=5):
    g = g.sort_values(T_COL)
    t = g[T_COL].to_numpy(float)
    y = g[RPM_COL].to_numpy(float)

    if len(t) < 3:
        return np.nan

    t0, t1 = float(t[0]), float(t[-1])
    ta = t0 + a * (t1 - t0)
    tb = t0 + b * (t1 - t0)

    mask = (t >= ta) & (t <= tb)
    yy = y[mask]

    # fallback si pocos puntos en ventana
    if len(yy) < min_pts:
        yy = y

    # cuantizamos para robustez
    yyq = np.round(yy / round_to) * round_to

    # modo
    vals, counts = np.unique(yyq, return_counts=True)
    if len(vals) == 0:
        return np.nan

    mode_val = float(vals[np.argmax(counts)])
    mode_count = int(np.max(counts))
    frac = float(mode_count / len(yyq)) if len(yyq) else np.nan
    return mode_val, mode_count, frac, len(yyq)

# ============================
# compute RPM_STABLE por PWM
# ============================
rows = []

for pwm, g_pwm in m.groupby(PWM_COL):
    pwm_int = int(pwm)

    # si hay varias runs, calculamos modo por run y luego agregamos (mediana)
    per_run = []
    for rid, g_run in g_pwm.groupby(RUN_COL):
        out = plateau_mode_rpm(g_run, round_to=ROUND_TO, a=PLATEAU_A, b=PLATEAU_B, min_pts=MIN_PTS_WINDOW)
        if isinstance(out, tuple) and len(out) == 4:
            mode_val, mode_count, frac, nwin = out
            per_run.append((rid, mode_val, mode_count, frac, nwin))

    if len(per_run) == 0:
        stable = np.nan
        n_runs = 0
        frac_med = np.nan
        nwin_med = np.nan
    else:
        # valor estable: mediana de los modos de cada run (robusto)
        stable = float(np.nanmedian([x[1] for x in per_run]))
        n_runs = len(per_run)
        frac_med = float(np.nanmedian([x[3] for x in per_run]))
        nwin_med = float(np.nanmedian([x[4] for x in per_run]))

    rows.append({
        "pwm": pwm_int,
        "rpm_stable": stable,
        "n_runs": n_runs,
        "plateau_mode_frac_median": frac_med,   # qu tan dominante es el modo en plateau
        "n_pts_plateau_median": nwin_med
    })

rpm_stable_df = pd.DataFrame(rows).sort_values("pwm").reset_index(drop=True)

# diccionario final
RPM_STABLE = dict(zip(rpm_stable_df["pwm"], rpm_stable_df["rpm_stable"]))

rpm_stable_df
```

```python
import numpy as np
import pandas as pd

PWM_COL = "pwm"
RUN_COL = "run_id"
T_COL   = "t_rel_s"
RPM_COL = "rpm_f" if "rpm_f" in m.columns else "rpm"

def _interp_cross_time(t1, y1, t2, y2, level):
    """Tiempo en el que la recta (t1,y1)-(t2,y2) cruza 'level'. Asume que level est entre y1 y y2."""
    if y2 == y1:
        return float(t2)  # plano: devolvemos extremo derecho (da igual aqu)
    a = (level - y1) / (y2 - y1)
    return float(t1 + a * (t2 - t1))

def first_crossing_time(t, y, level):
    """Primer cruce (en el tiempo) de y con 'level' (subiendo o bajando), interpolado."""
    t = np.asarray(t, float)
    y = np.asarray(y, float)
    for i in range(1, len(y)):
        y0, y1 = y[i-1], y[i]
        if (y0 - level) == 0:
            return float(t[i-1])
        # hay cruce si estn a distinto lado o uno justo en level
        if (y0 - level) * (y1 - level) <= 0:
            # si el segundo ya es exactamente level
            if y1 == level:
                return float(t[i])
            # si el primero ya es level
            if y0 == level:
                return float(t[i-1])
            return _interp_cross_time(t[i-1], y0, t[i], y1, level)
    return np.nan

def last_crossing_time(t, y, level):
    """ltimo cruce (en el tiempo) de y con 'level' (subiendo o bajando), interpolado."""
    t = np.asarray(t, float)
    y = np.asarray(y, float)
    for i in range(len(y)-1, 0, -1):
        y0, y1 = y[i-1], y[i]
        if (y1 - level) == 0:
            return float(t[i])
        if (y0 - level) * (y1 - level) <= 0:
            if y1 == level:
                return float(t[i])
            if y0 == level:
                return float(t[i-1])
            return _interp_cross_time(t[i-1], y0, t[i], y1, level)
    return np.nan

def compute_segments_for_run(g, rpm_stable, tol=0.0):
    """
    Devuelve:
    - t0, t_end
    - t_hit: primer instante en el que alcanza rpm_stable (fin aceleracin)
    - t_leave: ltimo instante en el que est en rpm_stable (inicio deceleracin)
    tol: si quieres banda alrededor del estable (ej: tol=50 rpm), se usa como level superior/inferior.
         Si tol=0, cruces exactos con rpm_stable.
    """
    g = g.sort_values(T_COL)
    t = g[T_COL].to_numpy(float)
    y = g[RPM_COL].to_numpy(float)

    t0, t_end = float(t[0]), float(t[-1])

    # Para hacerlo ms realista con ruido, permite banda:
    # - "llegar al estable" => cruzar level_up = rpm_stable - tol (y luego consideras ya dentro)
    # - "salir del estable" => cruzar level_down = rpm_stable - tol al bajar
    # Pero como t pediste "llegar al valor estable", por defecto tol=0.
    level = float(rpm_stable)

    t_hit = first_crossing_time(t, y, level)
    t_leave = last_crossing_time(t, y, level)

    # sanity: si por ruido el ltimo cruce sale antes que el primero, lo dejamos pero avisamos con NaN
    if np.isfinite(t_hit) and np.isfinite(t_leave) and t_leave < t_hit:
        t_leave = np.nan

    return {
        "t0": t0,
        "t_end": t_end,
        "rpm_stable": level,
        "t_hit": t_hit,
        "t_leave": t_leave,
        "t_acc": (t_hit - t0) if np.isfinite(t_hit) else np.nan,
        "t_dec": (t_end - t_leave) if np.isfinite(t_leave) else np.nan,
        "t_plateau": (t_leave - t_hit) if (np.isfinite(t_hit) and np.isfinite(t_leave)) else np.nan,
    }

# =========================
# Ejecutar para todas las runs
# =========================
rows = []
for (pwm, rid), g in m.groupby([PWM_COL, RUN_COL]):
    pwm = int(pwm)
    if pwm not in RPM_STABLE or pd.isna(RPM_STABLE[pwm]):
        continue

    out = compute_segments_for_run(g, RPM_STABLE[pwm], tol=0.0)
    out.update({PWM_COL: pwm, RUN_COL: int(rid), "n_samples": len(g)})
    rows.append(out)

segments = pd.DataFrame(rows).sort_values([PWM_COL, RUN_COL]).reset_index(drop=True)

segments
```

```python
import numpy as np
import pandas as pd

PWM_COL = "pwm"
RUN_COL = "run_id"
T_COL   = "t_rel_s"
RPM_COL = "rpm_f" if "rpm_f" in m.columns else "rpm"

# ---------------------------------------------------------
# Cruce interpolado (primer cruce dentro de una ventana)
# ---------------------------------------------------------
def _interp_cross_time(t1, y1, t2, y2, level):
    if y2 == y1:
        return float(t2)
    a = (level - y1) / (y2 - y1)
    return float(t1 + a * (t2 - t1))

def first_crossing_time(t, y, level):
    t = np.asarray(t, float)
    y = np.asarray(y, float)
    for i in range(1, len(y)):
        y0, y1 = y[i-1], y[i]
        if y0 == level:
            return float(t[i-1])
        if (y0 - level) * (y1 - level) <= 0:
            if y1 == level:
                return float(t[i])
            return _interp_cross_time(t[i-1], y0, t[i], y1, level)
    return np.nan

# ---------------------------------------------------------
# Mapa pwm -> rpm_stable
# ---------------------------------------------------------
if "RPM_STABLE" in globals() and isinstance(RPM_STABLE, dict):
    rpm_stable_map = {int(k): float(v) for k, v in RPM_STABLE.items() if pd.notna(v)}
else:
    stable_table = stable_table if "stable_table" in globals() else None
    if stable_table is None:
        raise NameError("No encuentro RPM_STABLE ni stable_table. Necesito pwm->rpm_stable.")
    rpm_stable_map = dict(zip(stable_table["pwm"].astype(int), stable_table["rpm_stable"].astype(float)))

# Seales objetivo (las que quieres)
PWM_LIST = [1200, 1300, 1400, 1500, 1600, 1800, 2000]
PWM_LIST = [p for p in PWM_LIST if p in rpm_stable_map]

# ---------------------------------------------------------
# Cruces en ACELERACIN para TODAS las seales
# ventana: [t0, t_hit]
# ---------------------------------------------------------
all_rows = []

for pwm_src in PWM_LIST:
    g_src = m[m[PWM_COL] == pwm_src]
    if g_src.empty:
        continue

    # en aceleracin solo tiene sentido cruzar hacia arriba hasta el propio pwm_src
    targets = [p for p in PWM_LIST if rpm_stable_map[p] <= rpm_stable_map[pwm_src]]

    for rid, g in g_src.groupby(RUN_COL):
        g = g.sort_values(T_COL)
        t = g[T_COL].to_numpy(float)
        y = g[RPM_COL].to_numpy(float)

        seg_row = segments[(segments[PWM_COL] == pwm_src) & (segments[RUN_COL] == int(rid))]
        if seg_row.empty:
            continue

        t0   = float(seg_row["t0"].iloc[0])
        tHit = float(seg_row["t_hit"].iloc[0])

        mask = (t >= t0) & (t <= tHit)
        tA, yA = t[mask], y[mask]
        if len(tA) < 2:
            continue

        row = {"pwm_src": int(pwm_src), "run_id": int(rid), "t0": t0, "t_hit": tHit}
        for p in targets:
            lvl = rpm_stable_map[p]
            row[f"t_cross_{p}"] = first_crossing_time(tA, yA, lvl)
        all_rows.append(row)

cross_acc_all = pd.DataFrame(all_rows).sort_values(["pwm_src", "run_id"]).reset_index(drop=True)
display(cross_acc_all)

# ---------------------------------------------------------
# Convertir a t entre seales (ACELERACIN) con medias
# dt(from->to) = mean(t_cross_to) - mean(t_cross_from)
# reposo: t_cross_0 = 0
# ---------------------------------------------------------
M_acc = pd.DataFrame(index=[0] + PWM_LIST, columns=[0] + PWM_LIST, dtype=float)

for pwm_src in PWM_LIST:
    sub = cross_acc_all[cross_acc_all["pwm_src"] == pwm_src]
    if sub.empty:
        continue

    # medias de cruces disponibles en esta seal
    mean_cross = {}
    for p in PWM_LIST:
        col = f"t_cross_{p}"
        if col in sub.columns and sub[col].notna().any():
            mean_cross[p] = float(sub[col].mean())

    mean_cross_with0 = {0: 0.0, **mean_cross}

    # para esta seal, podemos estimar tiempos hasta "pwm_src" (y desde 0)
    # guardamos en matriz global: filas=from, cols=to
    available = [0] + sorted([p for p in mean_cross.keys()], key=lambda x: rpm_stable_map[x])

    for a in available:
        for b in available:
            if b <= a:
                continue
            ta = mean_cross_with0.get(a, np.nan)
            tb = mean_cross_with0.get(b, np.nan)
            if np.isfinite(ta) and np.isfinite(tb):
                M_acc.loc[a, b] = tb - ta

display(M_acc)
```

```python
import numpy as np
import pandas as pd

PWM_COL = "pwm"
RUN_COL = "run_id"
T_COL   = "t_rel_s"
RPM_COL = "rpm_f" if "rpm_f" in m.columns else "rpm"

# ---------------------------------------------------------
# Cruce interpolado
# ---------------------------------------------------------
def _interp_cross_time(t1, y1, t2, y2, level):
    if y2 == y1:
        return float(t2)
    a = (level - y1) / (y2 - y1)
    return float(t1 + a * (t2 - t1))

def last_crossing_time(t, y, level):
    """
    ltimo cruce (en el tiempo) de y con 'level' dentro de la ventana.
    til para deceleracin: al bajar, pasas por niveles cada vez menores.
    """
    t = np.asarray(t, float)
    y = np.asarray(y, float)
    for i in range(len(y)-1, 0, -1):
        y0, y1 = y[i-1], y[i]
        if y1 == level:
            return float(t[i])
        if (y0 - level) * (y1 - level) <= 0:
            if y0 == level:
                return float(t[i-1])
            return _interp_cross_time(t[i-1], y0, t[i], y1, level)
    return np.nan

# ---------------------------------------------------------
# Mapa pwm -> rpm_stable
# ---------------------------------------------------------
if "RPM_STABLE" in globals() and isinstance(RPM_STABLE, dict):
    rpm_stable_map = {int(k): float(v) for k, v in RPM_STABLE.items() if pd.notna(v)}
else:
    stable_table = stable_table if "stable_table" in globals() else None
    if stable_table is None:
        raise NameError("No encuentro RPM_STABLE ni stable_table. Necesito pwm->rpm_stable.")
    rpm_stable_map = dict(zip(stable_table["pwm"].astype(int), stable_table["rpm_stable"].astype(float)))

# Seales objetivo
PWM_LIST = [1200, 1300, 1400, 1500, 1600, 1800, 2000]
PWM_LIST = [p for p in PWM_LIST if p in rpm_stable_map]

# ---------------------------------------------------------
# Cruces en DECELERACIN para TODAS las seales
# ventana: [t_leave, t_end]
# ---------------------------------------------------------
all_rows = []

for pwm_src in PWM_LIST:
    g_src = m[m[PWM_COL] == pwm_src]
    if g_src.empty:
        continue

    # en deceleracin cruzar hacia abajo niveles <= a su propio pwm_src
    targets = [p for p in PWM_LIST if rpm_stable_map[p] <= rpm_stable_map[pwm_src]]

    for rid, g in g_src.groupby(RUN_COL):
        g = g.sort_values(T_COL)
        t = g[T_COL].to_numpy(float)
        y = g[RPM_COL].to_numpy(float)

        seg_row = segments[(segments[PWM_COL] == pwm_src) & (segments[RUN_COL] == int(rid))]
        if seg_row.empty:
            continue

        tLeave = float(seg_row["t_leave"].iloc[0])
        tEnd   = float(seg_row["t_end"].iloc[0])

        if not np.isfinite(tLeave):
            continue

        mask = (t >= tLeave) & (t <= tEnd)
        tD, yD = t[mask], y[mask]
        if len(tD) < 2:
            continue

        row = {"pwm_src": int(pwm_src), "run_id": int(rid), "t_leave": tLeave, "t_end": tEnd}
        for p in targets:
            lvl = rpm_stable_map[p]
            row[f"t_cross_{p}"] = last_crossing_time(tD, yD, lvl)
        all_rows.append(row)

cross_dec_all = pd.DataFrame(all_rows).sort_values(["pwm_src", "run_id"]).reset_index(drop=True)
display(cross_dec_all)

# ---------------------------------------------------------
# Convertir a t entre seales (DECELERACIN) con medias
# dt(from->to) = mean(t_cross_to) - mean(t_cross_from)
# aqu: to < from
# ---------------------------------------------------------
M_dec = pd.DataFrame(index=[0] + PWM_LIST, columns=[0] + PWM_LIST, dtype=float)

for pwm_src in PWM_LIST:
    sub = cross_dec_all[cross_dec_all["pwm_src"] == pwm_src]
    if sub.empty:
        continue

    mean_cross = {}
    for p in PWM_LIST:
        col = f"t_cross_{p}"
        if col in sub.columns and sub[col].notna().any():
            mean_cross[p] = float(sub[col].mean())

    # en esta seal podemos tener cruces para p <= pwm_src
    available = sorted(mean_cross.keys(), key=lambda x: rpm_stable_map[x], reverse=True)

    for a in available:
        for b in available:
            if b >= a:
                continue
            ta = mean_cross.get(a, np.nan)
            tb = mean_cross.get(b, np.nan)
            if np.isfinite(ta) and np.isfinite(tb):
                M_dec.loc[a, b] = tb - ta

display(M_dec)
```

```python
import numpy as np
import pandas as pd

# =========================================================
# COMBINAR M_acc (arriba) + M_dec (abajo) + 0<->PWM desde segments
# =========================================================

PWM_AXIS = [0, 1200, 1300, 1400, 1500, 1600, 1800, 2000]

# --- matriz final ---
M_full = pd.DataFrame(index=PWM_AXIS, columns=PWM_AXIS, dtype=float)

# 1) copiar aceleracin (tringulo superior)
for i in PWM_AXIS:
    for j in PWM_AXIS:
        if (i != j) and (j > i):
            if "M_acc" in globals() and (i in M_acc.index) and (j in M_acc.columns):
                M_full.loc[i, j] = M_acc.loc[i, j]

# 2) copiar deceleracin (tringulo inferior)
for i in PWM_AXIS:
    for j in PWM_AXIS:
        if (i != j) and (j < i):
            if "M_dec" in globals() and (i in M_dec.index) and (j in M_dec.columns):
                M_full.loc[i, j] = M_dec.loc[i, j]

# 3) completar 0->PWM (aceleracin desde reposo) con segments.t_acc
#    completar PWM->0 (deceleracin a reposo) con segments.t_dec
if "segments" not in globals():
    raise NameError("No encuentro 'segments' (tabla con t_acc y t_dec). Ejecuta antes el clculo de segmentos.")

acc0 = segments.groupby("pwm")["t_acc"].mean()  # media si hubiera varias runs
dec0 = segments.groupby("pwm")["t_dec"].mean()

for p in PWM_AXIS:
    if p == 0:
        continue
    if p in acc0.index and np.isfinite(acc0.loc[p]):
        M_full.loc[0, p] = float(acc0.loc[p])
    if p in dec0.index and np.isfinite(dec0.loc[p]):
        M_full.loc[p, 0] = float(dec0.loc[p])

# diagonal NaN
np.fill_diagonal(M_full.values, np.nan)

display(M_full)

# =========================================================
# (opcional) Heatmap de la matriz completa
# =========================================================
import matplotlib.pyplot as plt

A = M_full.to_numpy(float)

plt.figure(figsize=(9, 7))
im = plt.imshow(A, origin="lower")
plt.colorbar(im, label="Tiempo transicin [s]")
plt.xticks(range(len(PWM_AXIS)), PWM_AXIS)
plt.yticks(range(len(PWM_AXIS)), PWM_AXIS)
plt.xlabel("PWM objetivo")
plt.ylabel("PWM inicial")
plt.title("Matriz completa de tiempos de transicin (M_acc arriba, M_dec abajo)")

for i in range(len(PWM_AXIS)):
    for j in range(len(PWM_AXIS)):
        val = A[i, j]
        if np.isfinite(val):
            plt.text(j, i, f"{val:.3f}", ha="center", va="center", fontsize=8)

plt.plot([-0.5, len(PWM_AXIS)-0.5], [-0.5, len(PWM_AXIS)-0.5], "k--", linewidth=1)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd

# -----------------------------
# Inputs
# -----------------------------
pwm_to_v = {
    1200: 2.3,
    1300: 3.7,
    1400: 5.2,
    1500: 6.8,
    1600: 8.8,
    1800: 11.1,
    2000: 12.4,
}
PWM_AXIS = [0, 1200, 1300, 1400, 1500, 1600, 1800, 2000]

# Build velocity axis (same ordering as matrix axes)
v_axis = [0.0] + [float(pwm_to_v[p]) for p in PWM_AXIS[1:]]
idx = list(range(len(v_axis)))  # indices in matrix

# Matrix in the PWM ordering
A = M_full.loc[PWM_AXIS, PWM_AXIS].to_numpy(float)

# -----------------------------
# Helper: pick upper triangle = acceleration, lower triangle = deceleration
# -----------------------------
def get_acc_values(A):
    vals = []
    n = A.shape[0]
    for i in range(n):
        for j in range(n):
            if j > i and np.isfinite(A[i, j]):
                vals.append(A[i, j])
    return np.array(vals, float)

def get_dec_values(A):
    vals = []
    n = A.shape[0]
    for i in range(n):
        for j in range(n):
            if i > j and np.isfinite(A[i, j]):
                vals.append(A[i, j])
    return np.array(vals, float)

acc_all = get_acc_values(A)
dec_all = get_dec_values(A)

# -----------------------------
# 1) Neighbor steps (one-level transitions)
# -----------------------------
acc_step = np.array([A[i, i+1] for i in range(len(v_axis)-1) if np.isfinite(A[i, i+1])], float)
dec_step = np.array([A[i+1, i] for i in range(len(v_axis)-1) if np.isfinite(A[i+1, i])], float)

# -----------------------------
# 2) Full range extreme-to-extreme
# -----------------------------
i_lo, i_hi = 1, len(v_axis)-1  # from 2.3 m/s to 12.4 m/s (exclude 0.0 by default)
t_range_acc = A[i_lo, i_hi]
t_range_dec = A[i_hi, i_lo]

# If you want to include rest (0.0) as low end instead:
t_from_rest_to_hi = A[0, i_hi]
t_from_hi_to_rest = A[i_hi, 0]

# -----------------------------
# 3) Frequency limits for toggling between two conditions
# -----------------------------
def fmax_toggle(t_up, t_down):
    if not (np.isfinite(t_up) and np.isfinite(t_down)) or (t_up + t_down) <= 0:
        return np.nan
    return 1.0 / (t_up + t_down)

fmax_range = fmax_toggle(t_range_acc, t_range_dec)
fmax_rest_hi = fmax_toggle(t_from_rest_to_hi, t_from_hi_to_rest)

# Recommended "working frequency" with margin
MARGIN = 0.3
fwork_range = MARGIN * fmax_range
fwork_rest_hi = MARGIN * fmax_rest_hi

# -----------------------------
# Summary table (nice for report)
# -----------------------------
def stats(x):
    x = np.asarray(x, float)
    x = x[np.isfinite(x)]
    if len(x) == 0:
        return {"n": 0, "median": np.nan, "p95": np.nan, "max": np.nan}
    return {
        "n": len(x),
        "median": float(np.median(x)),
        "p95": float(np.percentile(x, 95)),
        "max": float(np.max(x)),
    }

summary = pd.DataFrame([
    {"metric": "Neighbor step (acc) [s]", **stats(acc_step)},
    {"metric": "Neighbor step (dec) [s]", **stats(dec_step)},
    {"metric": "All acc transitions [s]", **stats(acc_all)},
    {"metric": "All dec transitions [s]", **stats(dec_all)},
])

display(summary)

print("\n--- RANGE (2.3  12.4 m/s) ---")
print(f"t_range_acc = {t_range_acc:.3f} s")
print(f"t_range_dec = {t_range_dec:.3f} s")
print(f"f_max(range)   {fmax_range:.3f} Hz")
print(f"f_work(range)  {fwork_range:.3f} Hz  (margin factor {MARGIN})")

print("\n--- REST  12.4 m/s ---")
print(f"t_rest_to_hi = {t_from_rest_to_hi:.3f} s")
print(f"t_hi_to_rest = {t_from_hi_to_rest:.3f} s")
print(f"f_max(rest-hi)   {fmax_rest_hi:.3f} Hz")
print(f"f_work(rest-hi)  {fwork_rest_hi:.3f} Hz  (margin factor {MARGIN})")
```
